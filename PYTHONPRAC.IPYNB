chains

```python
# simple chains sequential and custom chains 
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

chat= ChatOpenAI(model='gpt-3.5-turbo',temperature=0.5,max_tokens=1024)# model initialization 
template = ''' you are an experianced virologist write and few sentences about the following {virus} in {language}'''

Promp= PromptTemplate(input_variables=['virus','language'], template=template)


chain=LLMChain(llm=chat,prompt=Promp)

output=chain.run({
    'virus':'ebola',
    'language':"TELUGU"
})# THIS IS DONE IN  CASE OF MULTIPLE INPUT VARIABLES IN THE PROMPT TEMPLATE 

#output=chain.run('ebola') # this is case of  single input variable in prompt  template 
output
```

```python
#sequential chains 
# it is used in case of mutiple sequential questions ( first questions aswer is the quesation for 2nd one)

from langchain.chat_models import ChatOpenAI
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain,SimpleSequentialChain


llm1= OpenAI(model='text-davinci-003',temperature=0.7, max_tokens=1024)

prompt1= PromptTemplate(
    input_variables=['concept'],
    template=''' you are an experienced scientist and python programmer write a function that impliments this {concept}'''
    
)

chain=LLMChain(llm=llm1,prompt=prompt1)



llm2=ChatOpenAI(model='gpt-3.5-turbo', temperature=0.7)

prompt2= PromptTemplate(
    input_variables=['function'],
    template=''' Given python function {function} , describe it as detailed as possible'''
)
chain2=LLMChain(llm=llm2,prompt=prompt2)

overall_chain=SimpleSequentialChain(chains=[chain,chain2], verbose=True)

output= overall_chain.run('lenear regression')

```

```python
#agents are the way to provide llms the ability to do math , search out side world 
# for math problem we can use python agent 

from langchain.agents.agent_toolkits import create_python_agent
from langchain.tools.python.tool import PythonREPLTool
from langchain.llms import OpenAI

llm=OpenAI(temperature=0.7)
agent_executioner = create_python_agent(
    llm=llm,
    tool=PythonREPLTool(),
    verbose=True
)

agent_executioner.run('calculate the squere root of the factorial of 20 and display with 4 decimals')
```

```python
import os
from dotenv import load_dotenv,find_dotenv
load_dotenv(find_dotenv(),override=True)


import pinecone

pinecone.init(
    api_key=os.environ.get('PINECONE_API_KEY'),
    environment=os.environ.get('PINECONE_ENV')
   
)
```

pinecone indexes

```python
pinecone.list_indexes()
```

```python
# if no indexes are found or you want to create an new index
index_name = 'medium-blogs-embedding-index'
if index_name not in pinecone.list_indexes():
    print(f"creating new index{index_name}")
    pinecone.create_index(
        index_name, dimension=1536, metric="cosine", pods=1, pod_type="p1.x2"
    )
    print('done')
else:
    print(f'index {index_name} already exits ')
```

```python
#index discription
pinecone.describe_index(index_name)#describes the index 
```

```python
# deleting the index
index_name='langchain-index'
if index_name in pinecone.list_indexes():
    print(f'deletting index {index_name}')
    pinecone.delete_index(index_name)
    print('done')
else:
    print(f'index {index_name} does not exit ')
```

```python
# creating random vectors with 5 different ids
import random
vectors=[[random.random() for _ in range(1536)]for v in range(5)]
vectors
ids=list('abcde')
```

```python
# inserting vectors and ids to pinecone 
index_name = 'medium-blogs-embedding-index'
index = pinecone.Index(index_name)
index.upsert(vectors=zip(ids,vectors))# zip puts vecotors in ids (name spaces)
```

```python
#updating pinecone 

index.upsert(vectors=[('c', [0.3]*1536)])
```

```python
#fetching vectors 
index_name = 'medium-blogs-embedding-index'
index = pinecone.Index(index_name)
index.fetch(ids=['a','b'])
```

```python
#delete vectors by id 
index.delete(ids=['c','b'])
```

```python
index.describe_index_stats()
```
